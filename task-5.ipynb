```python
pairplot_cols = ['Survived', 'Pclass', 'Age', 'Fare', 'SibSp', 'Parch']
pairplot_df = df_app_data[pairplot_cols].dropna()
```

```python
sns.pairplot(pairplot_df, hue='Survived', palette='Set2')
plt.suptitle("Pairplot of Titanic Features by Survival", y=1.02)
plt.show()
```

```python
correlation = df_app_data[pairplot_cols].corr()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Titanic Dataset")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df_app_data['Age'].plot.hist(bins=30, edgecolor='black')
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Survived', y='Age', data=df_app_data)
plt.title("Age Distribution by Survival")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Pclass', y='Fare', data=df_app_data)
plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df_app_data, palette='Set1')
plt.title("Fare vs Age Colored by Survival")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
